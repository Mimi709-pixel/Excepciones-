```python
# Gestión de Grupo - Mundial de Fútbol
# Librerías usadas
import pandas as pd  # para mostrar la tabla bonita

# --- 1. DEFINIR EQUIPOS DEL GRUPO ---
equipos_nombres = ["Colombia", "Brasil", "Argentina", "Uruguay"]

# --- 2. INICIALIZAR ESTADÍSTICAS ---
def crear_tabla(equipos):
    tabla = {}
    for equipo in equipos:
        tabla[equipo] = {
            "PJ": 0,  # Partidos jugados
            "PG": 0,  # Partidos ganados
            "PE": 0,  # Partidos empatados
            "PP": 0,  # Partidos perdidos
            "GF": 0,  # Goles a favor
            "GC": 0,  # Goles en contra
            "Pts": 0  # Puntos
        }
    return tabla

# --- 3. REGISTRAR UN PARTIDO ---
def registrar_partido(tabla):
    print("\n--- Equipos disponibles ---")
    for i, equipo in enumerate(tabla.keys()):
        print(f"  {i+1}. {equipo}")

    equipo_a = input("Nombre del equipo local: ").strip()
    equipo_b = input("Nombre del equipo visitante: ").strip()

    if equipo_a not in tabla or equipo_b not in tabla:
        print("Uno o ambos equipos no están en el grupo.")
        return

    goles_a = int(input(f"Goles de {equipo_a}: "))
    goles_b = int(input(f"Goles de {equipo_b}: "))

    # Actualizar PJ y goles
    tabla[equipo_a]["PJ"] += 1
    tabla[equipo_b]["PJ"] += 1
    tabla[equipo_a]["GF"] += goles_a
    tabla[equipo_a]["GC"] += goles_b
    tabla[equipo_b]["GF"] += goles_b
    tabla[equipo_b]["GC"] += goles_a

    # Calcular resultado
    if goles_a > goles_b:
        print(f"Gana {equipo_a}")
        tabla[equipo_a]["PG"] += 1
        tabla[equipo_a]["Pts"] += 3
        tabla[equipo_b]["PP"] += 1
    elif goles_b > goles_a:
        print(f"Gana {equipo_b}")
        tabla[equipo_b]["PG"] += 1
        tabla[equipo_b]["Pts"] += 3
        tabla[equipo_a]["PP"] += 1
    else:
        print("Empate")
        tabla[equipo_a]["PE"] += 1
        tabla[equipo_a]["Pts"] += 1
        tabla[equipo_b]["PE"] += 1
        tabla[equipo_b]["Pts"] += 1

# --- 4. MOSTRAR TABLA DE POSICIONES ---
def mostrar_tabla(tabla):
    print("\n TABLA DE POSICIONES")
    df = pd.DataFrame(tabla).T  # Transponer para que equipos sean filas
    df = df.sort_values(by=["Pts", "GF"], ascending=False)
    df.index.name = "Equipo"
    print(df.to_string())

# --- 5. MENÚ PRINCIPAL ---
tabla = crear_tabla(equipos_nombres)

while True:
    print("\n======= MUNDIAL - GRUPO A =======")
    print("1. Registrar partido")
    print("2. Ver tabla de posiciones")
    print("3. Salir")

    opcion = input("Elige una opción: ").strip()

    if opcion == "1":
        registrar_partido(tabla)
    elif opcion == "2":
        mostrar_tabla(tabla)
    elif opcion == "3":
        print("¡Hasta luego!")
        break
    else:
        print("Opción no válida.")
```
